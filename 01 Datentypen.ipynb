## Kommentare schreiben in Python

Es ist immer eine gute Idee, nicht nur Code zu schreiben, sondern deinem Code auch Kommentare hinzuzufügen. Es wird anderen helfen zu verstehen, was du versucht hast zu erreichen (der Grund, warum du einen bestimmten Ausschnitt des Codes geschrieben hast). Dies hilft nicht nur **anderen Leuten**, deinen Code zu verstehen, es kann auch als Erinnerung dienen **für dich**, wenn du Wochen oder Monate später darauf zurückkommst.

Um Kommentare in Python zu schreiben, benutze das Rautezeichen `#`, bevor du deinen Kommentar schreibst. Wenn du deinen Code ausführst, ignoriert Python alles, was hinter dem `#` in einer bestimmten Zeile steht.

```
# Übung im Schreiben von Kommentaren

print('Hallo, Python!') # Diese Zeile gibt eine Zeichenkette aus.
# print('Hi')
```

Nach der Ausführung der obigen Zelle solltest du beachten, dass `Diese Zeile gibt eine Zeichenkette aus` nicht in der Ausgabe erschienen ist, da es sich um einen Kommentar handelte (und somit von Python ignoriert wurde).

Die zweite Zeile wurde auch nicht ausgeführt, weil `print('Hi')` ebenfalls mit dem Nummernzeichen (`#`) versehen war! Da es sich hierbei nicht um einen erklärenden Kommentar des Programmierers, sondern um eine tatsächliche Codezeile handelt, sagt man, dass der Programmierer diese zweite Zeile des Codes **auskommentiert** hat, da er ihn vielleicht ein anderes mal doch benötigt und ihn nicht ganz löschen will.

## Fehler (Errors) in Python

Jeder macht Fehler. Für viele Arten von Fehlern wird Python dir sagen, dass du einen Fehler gemacht hast, indem es dir eine Fehlermeldung ausgibt. Es ist wichtig, die Fehlermeldungen sorgfältig zu lesen, um wirklich zu verstehen, wo du einen Fehler gemacht hast und wie du ihn korrigieren kannst.

Zum Beispiel, wenn du `print` als `frint` schreibst, wird Python eine Fehlermeldung anzeigen. Probieren es aus:

```
# Print falsch geschrieben erzeugt eine Fehlermeldung

frint("Hallo, Python!")
```

Die Fehlermeldung sagt dir in diesem Fall: 

*   wo (hier: in Zeile 2) der Fehler aufgetreten ist, und
*   welcher Art von Fehler es war (hier: *NameError*)

Hier hat Python hat versucht, die Funktion `frint` auszuführen, konnte aber nicht feststellen, was `frint` sein soll, da es sich um keine der eingebauten Funktionen handelt und `frint` auch nicht vorher von uns als Funktion definiert wurde.

Du wirst feststellen, dass, wenn wir einen anderen Fehler machen, indem wir z.B. vergessen, die Zeichenkette mit einem zweiten Anführungszeichen zu schließen, wir einen anderen Fehler erhalten (z.B. einen `SyntaxError`).

## Erkennt Python deinen Fehler, bevor es deinen Code ausführt?

Python ist eine so genannte *interpretierte Sprache* im Gegensatz zu *kompilierten* (übersetzten) Sprachen. Kompilierte Sprachen untersuchen dein gesamtes Programm wenn es in für den Computer verständlichen Code übersetzt ("kompiliert") wird. Solche Sprachen sind in der Lage, dich schon vor der Ausführung über eine ganze Reihe von Fehlern zu informieren. 

Im Gegensatz dazu interpretiert Python deinen Code Zeile für Zeile, während er ausgeführt wird. Python hört auf, das gesamte Programm auszuführen, wenn es auf einen Fehler stößt (es sei denn, der Fehler wird vom Programmierer vorausgesehen und abgefangen, ein fortgeschritteneres Thema, das wir vielleicht später in diesem Kurs behandeln werden).

## Umwandlung von einem Objekttyp in einen anderen

Du kannst den Typ eines Objekts in Python ändern. Man nennt das *Typecasting*.

Du kannst beispielsweise einen *Integer*-Wert in einen *Float*-Wert (z.B. 2 in 2.0) umwandeln (konvertieren).

Probieren wir das aus:

```
# Überprüfung, ob es sich um eine Ganzzahl (Integer) handelt.

type(2)
```

### Umwandlung (Casting) von einer Ganzzahl in eine Fließkommazahl
Wir möchten den Integerwert 2 in einen Floatwert umwandeln:

```
# Konvertiere 2 zu float

float(2)
```

```
# Konvertiere 2 (integer) zu float und überprüfe diesen Wert

type(float(2))
```

Wenn wir Integer in Float umwandeln, ändern wir nicht wirklich den Wert (d.h. die Genauigkeit) der Zahl. Aber bei der Umwandlung von Float in Integer könnten wir möglicherweise einige Informationen verlieren.

Wenn wir zum Beispiel den Float 1.1 nach Integer umwandeln, erhalten wir 1 und verlieren die Dezimalinformationen (d.h. 0.1):

```
# Das Casting von 1.1 in eine ganze Zahl führt zum Verlust von Informationen.

int(1.1)
```

Bei dieser Umwandlung wird nicht gerundet, sondern einfach alles was nach dem Komma kommt abgeschnitten. 

```
# Das Casting von 1.9 in eine ganze Zahl führt zum Verlust von Informationen.

int(1.9)
```

### Konvertierung von Strings in Integer oder Floats

```
# Konvertiere einen String in einen Integer-Wert.

int('1')
```

<p>Aber wenn du versuchst, dies mit einem String zu machen, der nicht perfekt mit einer Zahl übereinstimmt, wirst du einen Fehler erhalten. Versuche es einmal:</p>

## Boole'scher Datentyp

*Bool* ist ein weiterer wichtiger Typ in Python. Ein Objekt vom Typ *Bool* kann einen von zwei Werten annehmen: `True` oder `False` (Richtig oder Falsch bzw. Ja oder Nein bzw. 1 oder 0)


```
# Wert true

True
```

Beachtet, dass der Wert `True` einen Großbuchstaben "T" hat. Dasselbe gilt für `False` (d.h. du musst den Großbuchstaben "F" verwenden).

```
# Wert false

False
```

Wenn du Python aufforderst, den Typ eines booleschen Objekts anzuzeigen, zeigt es die Abkürzung `bool` an. 

```
# Typ von True

type(True)
```

```
# Typ von False

type(False)
```

Solche Bool-Werte kommen sehr häufig vor. Zum Beispiel gibt Python so das Ergebnis beim Vergleich zweier Zahlen zurück.

Wir überprüfen z.B. ob 2 kleiner als 5 ist:

```
# Stimmt es dass 2 kleiner als 5 ist?
2<5
```

Wir können auch boolesche Objekte in andere Datentypen umwandeln.

*   Wenn wir einen Boolean mit einem Wert von `True` in einen Integer oder Float umwandeln, erhalten wir 1. 
*   Wenn wir einen Boolean mit einem Wert von `False` nach Integer oder Float umwandeln, erhalten wir 0. 
*   Umgekehrt geht das genauso: wenn wir eine 1 nach Bool umwandeln, erhalten wir `True`. Und wenn wir eine 0 nach Bool umwandeln, erhalten wir `False`. 

Versuchen wir es mal: 

```
# Konvertieren von True zu int

int(True)
```

```
# Konvertiert 1 in boolean

bool(1)
```

```
# Konvertiert 0 in boolean

bool(0)
```

```
# True in float konvertieren

float(True)
```

## Variablen

Wie bei den meisten Programmiersprachen können wir Werte in *Variablen* speichern, damit wir sie später verwenden können. Zum Beispiel:

```
# Wert in Variable speichern

x = 43 + 60 + 16 + 41
```

Um den Wert von `x` in unserem Notebook zu sehen, können wir ihn einfach auf die letzte Zeile einer Zelle setzen:

```
# Ausgabe des Wertes in der Variable

x
```

Wir können auch Operationen mit `x` durchführen und das Ergebnis in einer neuen Variablen speichern:

```
# Verwendung einer anderen Variablen, um das Ergebnis der Operation zu speichern

y = x / 60
y             # Gibt den Wert von y aus
```

Wenn wir einen Wert in einer bestehenden Variablen speichern, überschreibt der neue Wert den vorherigen Wert:

```
# Überschreiben der Variablen mit einem neuen Wert

print(x)     # Ausgabe des aktuellen x Wertes
x = x / 60
x            # Ausgabe des neuen x-Wertes
```

Es ist empfehlenswert, sinnvolle Variablennamen zu verwenden, damit du und andere den Code leichter lesen und verstehen können:

```
# Bezeichne die Variablen sinnvoll

cd_gesamt_min = 43 + 42 + 57 # Gesamtlänge von 3 Musik-CDs in Minuten
cd_gesamt_min
```

```
# Bezeichne die Variablen sinnvoll

cd_gesamt_stunden = cd_gesamt_min / 60 # Gesamtlänge der Musik-CDs in Stunden
cd_gesamt_stunden
```

In den obigen Zellen haben wir die Länge von drei CDs in Minuten hinzugefügt und in `cd_gesamt_min` gespeichert. Wir haben den Wert dann durch 60 geteilt, um die Gesamtlänge `cd_gesamt_stunden` in Stunden zu berechnen.

Du kannst das auch alles auf einmal in einem einzigen Ausdruck erledigen, solange du die Länge der CDs in Klammern setzt, bevor du teilst. Das würde dann so aussehen.

```
# Komplizierterer Ausdruck

cd_gesamt_stunden = (43 + 42 + 57) / 60  # Gesamtstunden in einem einzigen Ausdruck
cd_gesamt_stunden
```
