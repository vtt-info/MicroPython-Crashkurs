# Was macht die Schildkröte?
Der untenstehende Code ist vielleicht noch etwas verwirrend, da eine sogenannte for-Schleife, die bisher noch nicht wirklich erklärt wurde, und der (nur in Google Colab) benötigte Befehl `initializeTurtle()` verwendet wird.

Macht euch keine Sorgen, wenn ihr diesen Code noch nicht so ganz versteht. Beobachtet einfach, was beim Ausführen passiert!

Was wird gezeichnet?

```
from ColabTurtle import *

initializeTurtle()   # Voreinstellung: speed = 4, Bildschirm = (800,500)
right(90)            # normalerweise schaut die Turtle anfangs nach rechts
#-----------------------------------------------------------------------

for i in range(2):
  forward(300)
  left(90)
  forward(150)
  left(90)

penup()
setx(0)
sety(0)
pendown()
color('blue')
goto(800,500)

```

# Die wichtigsten Befehle
Die wichtigsten Turtle-Befehle:

`initializeTurtle(Geschwindigkeit, (Breite,Höhe))` -  Legt die Malgeschwindigkeit fest und erstellt ein Fenster

`forward(Strecke)` - Die Schildkröte bewegt sich um "Strecke" nach vorne.

`backward(Strecke)` -	Die Schildkröte bewegt sich um "Strecke" rückwärts.

`left(Winkel)` -	Die Schildkröte dreht sich um "Winkel" nach links.

`right(Winkel)` -	Die Schildkröte dreht sich um "Winkel" nach rechts.

`penup()` -	Die Schildkröte hebt den Zeichenstift an (und schreibt nicht mehr).

`pendown()` -	Die Schildkröte senkt den Zeichenstift (und schreibt wieder).

`width(Breite)`	- setzt die Stiftdicke auf "Breite".

`color(Farbe)` -	setzt die Stiftfarbe auf "Farbe" (*'white', 'yellow', 'orange', 'red', 'green', 'blue', 'purple', 'grey', 'black'*)

`speed(Geschwindigkeit)`  -  ändert die Bewegungsgeschwindigkeit

`goto(x, y)`	-  Bewege die Schildkröte zur Position x, y.

`setx(x)`	-  Bewege die Schildkröte zur Position x, y bleibt gleich.

`sety(y)`	-  Bewege die Schildkröte zur Position y, x bleibt gleich.

`showturtle()`  -  Schildkröte ist sichtbar

`hideturtle()`  -  Schildkröte ist unsichtbar
