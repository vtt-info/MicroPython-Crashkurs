# String Operationen 

**Willkommen!** Dieses Notebook wird dir die Stringoperationen in der Programmiersprache Python erklären. Am Ende dieser Einheit wirst du die Grundlagen der Stringoperationen in Python kennen, einschließlich Indizierung, Escape-Sequenzen und Operatoren.

## Was sind Strings? 

Das folgende Beispiel zeigt einen String, der innerhalb von 2 Anführungszeichen steht:

```
# Anführungszeichen zum Festlegen von Strings verwenden

"Michael Jackson"
```

Wir können auch einfache Anführungszeichen verwenden:

```
# Verwendung von einfachen Anführungszeichen bei der Definition von Strings

'Michael Jackson'
```

Ein String kann eine Kombination aus Leerzeichen und Ziffern sein: 

```
# Ziffern und Leerzeichen in Strings

'1 2 3 4 5 6 '
```

Ein String kann auch eine Kombination von Sonderzeichen sein: 

```
# Sonderzeichen in Strings

'@#2_#]&*^%$'
```

Wir können unseren String mit der print-Anweisung ausgeben::

```
# Ausgabe eines Strings

print("Hallo!")
```

Wir können einen String einer Variablen zuweisen:


```
# String einer Variablen zuweisen

Name = "Herr March"
print(Name)
```

## Aneinanderfügen von Strings</h3>

Wir können Strings verketten oder kombinieren, indem wir die Additionssymbole verwenden, und das Ergebnis ist ein neuer String, der eine Kombination aus allen ist:


```
# Verkettung von zwei Strings

Bewunderung = Name + " ist der Beste von " + str(7)
Bewunderung

strHundename = "Beisso"
strHundealter = "4"
intMenschenjahre = int(strHundealter) * 7
print("Dein " + strHundealter +" Jahre alter Hund " +strHundename+" ist "+str(intMenschenjahre)+" Jahre alt in Menschenjahren.")
```

Um Werte eines Strings zu vervielfältigen, multiplizieren wir den String einfach mit der Anzahl der Male, die wir ihn vervielfältigen möchten. In diesem Fall ist die Zahl drei. 

Das Ergebnis ist ein neuer String, und dieser neue String besteht aus drei Kopien des ursprünglichen Strings:

```
# Den String 3 mal schreiben.

3 * "Michael Jackson. "
```

Du kannst einen neuen String erstellen, indem du ihn auf die ursprüngliche Variable setzt. Verknüpft mit einer neuen Zeichenkette entsteht ein neuer String, der von Michael Jackson zu "Michael Jackson ist der Beste" wechselt.


```
# Strings verketten

Name = "Michael Jackson"
Name = Name + " ist der Beste"
Name
```

## Escape-Sequenzen 

Backslashes `\\` bilden den Anfang von Escape-Sequenzen. Escape-Sequenzen stellen Zeichenketten dar, die  Sonderfunktionen ausführen. Beispielsweise repräsentiert "\n" eine neue Zeile. Die Ausgabe erfolgt durch eine neue Zeile dort wo "\n" steht:

```
# Escape-Sequenz für Zeilenumbrüche

print(" Michael\n Jackson \n ist der Beste" )
```

Ebenso stellt "\t" einen Tabulator dar: 

```
# Escape-Sequenz für Tabulatoren

print(" Michael Jackson \t ist der Beste" )
```

Wenn du einen Backslash in deinen String einfügen möchtest, verwende einen doppelten Backslash:

```
# Einfügen eines Backslashs in einen String

print(" Michael Jackson \\ ist der Beste" )
```

Wir können auch ein "r" vor die Zeichenkette setzen, um den Backslash anzuzeigen:

```
# r teilt Python mit, dass diese Zeichenkette als 'roher' String (englisch 'raw') angezeigt wird.

print(r" Michael Jackson \ ist der Beste" )
```

## String-Operationen 

Es gibt viele Methoden zum Bearbeiten von Zeichenketten in Python, die verwendet werden können, um Texte zu manipulieren.

Wir werden einige grundlegende Zeichenkettenoperationen auf Texte anwenden. 

Beginnen wir mit der Methode `upper`; diese Methode konvertiert Kleinbuchstaben in Großbuchstaben:

```
# Konvertiert alle Zeichen in Strings in Großbuchstaben.

A = "Thriller ist sein sechstes Studioalbum"
print("vor upper:", A)
B = A.upper()
print("nach upper:", B)
"KLEIN".lower()
```

Die Methode `replace` ersetzt ein Segment des Strings, d.h. einen Teilstring durch einen neuen String. Wir geben den Teil des Strings ein, den wir ändern möchten. Das zweite Argument ist, womit wir das Segment austauschen möchten, und das Ergebnis ist ein neuer String mit geändertem Segment: 


```
# Ersetzt den alten Teilstring durch den neuen Teilstring, wenn das Segment in dem String gefunden wurde.

A = "Michael Jackson ist der Beste"
B = A.replace('Michael', 'Janet')
B = B.replace('der', 'die')
B
```

Die Methode `find` findet einen Teilstring (Substring). Das Argument ist der Teilstring, den du finden möchtest, und das Ergebnis ist der erste Index mit dieser Sequenz. Wir können den Substring `jack` oder `el` finden.  

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%201/Images/StringsFind.png" width="600" align="center" />

```
# Finde den Teilstring in dem String. Nur der Index des ersten Teilstrings in der Zeichenkette ist das Ergebnis.

Name = "Michael Jackson"
Name.find('el')
```

```
# Finde den Substring im String.

Name.find('Jack')
```

Wenn sich der Substring nicht in der Zeichenkette befindet, dann ist das Ergebnis negativ. Beispielsweise ist die Zeichenkette 'Jasdfasdasdf' kein Substring:

```
# Wenn der Substring im String nicht gefunden wird.

Name.find('Jasdfasdasdf')
```
