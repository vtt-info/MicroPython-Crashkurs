## Escape-Sequenzen 

Backslashes `\\` bilden den Anfang von Escape-Sequenzen. Escape-Sequenzen stellen Zeichenketten dar, die  Sonderfunktionen ausf√ºhren. Beispielsweise repr√§sentiert "\n" eine neue Zeile. Die Ausgabe erfolgt durch eine neue Zeile dort wo "\n" steht:

```
# Escape-Sequenz f√ºr Zeilenumbr√ºche

print(" Michael\n Jackson \n ist der Beste" )
```

Ebenso stellt "\t" einen Tabulator dar: 

```
# Escape-Sequenz f√ºr Tabulatoren

print(" Michael Jackson \t ist der Beste" )
```

Wenn du einen Backslash in deinen String einf√ºgen m√∂chtest, verwende einen doppelten Backslash:

```
# Einf√ºgen eines Backslashs in einen String

print(" Michael Jackson \\ ist der Beste" )
```

Wir k√∂nnen auch ein "r" vor die Zeichenkette setzen, um den Backslash anzuzeigen:

```
# r teilt Python mit, dass diese Zeichenkette als 'roher' String (englisch 'raw') angezeigt wird.

print(r" Michael Jackson \ ist der Beste" )
```

## String-Operationen 

Es gibt viele Methoden zum Bearbeiten von Zeichenketten in Python, die verwendet werden k√∂nnen, um Texte zu manipulieren.

Wir werden einige grundlegende Zeichenkettenoperationen auf Texte anwenden. 

Beginnen wir mit der Methode `upper`; diese Methode konvertiert Kleinbuchstaben in Gro√übuchstaben:

```
# Konvertiert alle Zeichen in Strings in Gro√übuchstaben.

A = "Thriller ist sein sechstes Studioalbum"
print("vor upper:", A)
B = A.upper()
print("nach upper:", B)
"KLEIN".lower()
```

Die Methode `replace` ersetzt ein Segment des Strings, d.h. einen Teilstring durch einen neuen String. Wir geben den Teil des Strings ein, den wir √§ndern m√∂chten. Das zweite Argument ist, womit wir das Segment austauschen m√∂chten, und das Ergebnis ist ein neuer String mit ge√§ndertem Segment: 


```
# Ersetzt den alten Teilstring durch den neuen Teilstring, wenn das Segment in dem String gefunden wurde.

A = "Michael Jackson ist der Beste"
B = A.replace('Michael', 'Janet')
B = B.replace('der', 'die')
B
```

Die Methode `find` findet einen Teilstring (Substring). Das Argument ist der Teilstring, den du finden m√∂chtest, und das Ergebnis ist der erste Index mit dieser Sequenz. Wir k√∂nnen den Substring `jack` oder `el` finden.  

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%201/Images/StringsFind.png" width="600" align="center" />

```
# Finde den Teilstring in dem String. Nur der Index des ersten Teilstrings in der Zeichenkette ist das Ergebnis.

Name = "Michael Jackson"
Name.find('el')
```

```
# Finde den Substring im String.

Name.find('Jack')
```

Wenn sich der Substring nicht in der Zeichenkette befindet, dann ist das Ergebnis negativ. Beispielsweise ist die Zeichenkette 'Jasdfasdasdf' kein Substring:

```
# Wenn der Substring im String nicht gefunden wird.

Name.find('Jasdfasdasdf')
```
