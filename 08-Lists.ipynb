<h2 id="list">Lists</h2>

<h3 id="index">Indizierung</h3>

Wir werden uns Listen in Python ansehen. Eine Liste ist eine zusammengefügte Sammlung von verschiedenen Objekten wie ganzen Zahlen, Zeichenketten und anderen Elementen (dazu zählen auch andere Listen). Die Adresse jedes Elements innerhalb einer Liste wird als <b>Index</b> bezeichnet. Ein Index dient dem Zugriff und dem Verweis auf Elemente innerhalb einer Liste.

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsIndex.png" width="1000" />

Um eine Liste zu erstellen, gibst du die Liste in eckigen Klammern <b>[ ]</b> ein, mit dem Inhalt in der Klammer und getrennt durch Kommas. Wir probieren das gleich einmal aus!

```python
# Erstellung einer Liste

L = ["Michael Jackson", 10.1, 1982]
L
```

Bei einer Liste können wir eine negative und normale Indizierung verwenden:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%202/Images/ListsNeg.png" width="1000" />

```python
# Ausgabe von dem Element zu jedem Index

print('das gleiche Element mit negativer und positiver Indizierung:\n Postiv:',L[0],
'\n Negativ:' , L[-3]  )
print('das gleiche Element mit negativer und positiver Indizierung:\n Postiv:',L[1],
'\n Negativ:' , L[-2]  )
print('das gleiche Element mit negativer und positiver Indizierung:\n Postiv:',L[2],
'\n Negativ:' , L[-1]  )
```

<h3 id="content">Listeninhalt</h3>

Listen können Strings, Floats und Integerwerte enthalten. Wir können sogar andere Listen, Tupel und andere Datenstrukturen verschachteln. Für die Verschachtelung gelten die gleichen Indizierungsregeln:    


```python
# Beispielliste

["Michael Jackson", 10.1, 1982, [1, 2], ("A", 1)]
```

<h3 id="op">Listen-Operationen</h3>

Wir können die Methode <code>extend</code> verwenden, um neue Elemente zur Liste hinzuzufügen:

```python
# Anwendung von extend um der Liste Elemente hinzuzufügen

L = [ "Michael Jackson", 10.2]
L.extend(['pop', 10])
L
```

```python
L[0],L[1],L[2],L[3]    # beachte, dass L[2] den Wert 'pop' und L[3] den Wert 10 zurückgibt
```

Eine weitere ähnliche Methode ist <code>append</code>. Wenn wir <code>append</code> anstelle von <code>extend</code> anwenden, fügen wir ein Element zur Liste hinzu:

```python
# Anwendung von append um der Liste Elemente hinzuzufügen

L = [ "Michael Jackson", 10.2]
L.append(['pop', 10])
L
```

```python
L[2]   # L[2] ist eine Liste
```

Jedes Mal, wenn wir eine Methode anwenden, ändert sich die Liste. Wenn wir <code>extend</code> anwenden, fügen wir zwei neue Elemente zur Liste hinzu. Die Liste <code>L</code> wird dann durch Hinzufügen von zwei neuen Elementen geändert:

```python
# Anwendung von extend um der Liste Elemente hinzuzufügen

L = [ "Michael Jackson", 10.2]
L.extend(['pop', 10])
L
```

Listen sind veränderbar. Wir können beispielsweise das erste Element wie folgt ändern:

```python
# Ändern des Elements anhand des Indexes

A = ["disco", 10, 1.2]
print('Vor der Änderung:', A)
A[0] = 'hard rock'
print('Nach der Änderung:', A)
```

Wir können auch ein Element einer Liste mit dem Befehl <code>del</code> löschen:

```python
# Löschen von Elementen anhand des Indexes

print('Vor der Änderung:', A)
del(A[0])
print('Nach der Änderung:', A)
```
