<h2 id="loop">Schleifen (Loops)</h2>

Manchmal möchtest du einen bestimmten Vorgang mehrmals wiederholen. Wiederholte Ausführungen wie diese werden mit Hilfe von **Schleifen** durchgeführt. Wir werden uns zwei Arten von Schleifen ansehen, `for`- und `while`-Schleifen.

##Range()##

Bevor wir uns den Schleifen widmen, sollten wir das Objekt `range` besprechen. Es ist hilfreich, das Range-Objekt als geordnete Liste von zB Zahlen zu betrachten `[0,1,2,3,4,...]`. In eckigen Klammern werden in Python sogenannte Listen geschrieben. Jede Zahl in dieser Liste `[0,1,2,3,4,...]` entspricht auch dem sogenannten *Index*. Das ist die Stelle in der Liste, an der die Zahl zu finden ist. Python beginnt also beim Zählen nicht mit `1` sondern mit `0`.

Gehen wir zunächst vom einfachsten Fall aus. Wenn wir eine Zahlenreihe erzeugen möchten, die drei geordnete Elemente von 0 bis 2 enthält, verwenden wir einfach den folgenden Befehl:

```python
# Verwendung von range()

range(3)
```

##`for`-Schleifen##

Eine `for`-Schleife ermöglicht es dir, einen Codeblock mehrmals auszuführen. Du könntest das z. B. verwenden, wenn du jedes Element in einer Liste ausgeben möchtest.
   
Versuchen wir, eine `for`-Schleife zu verwenden, um alle Jahre auszugeben, die in der Liste `dates` enthalten sind:

Das kann folgendermaßen realisiert werden:

```python
# Beispiel einer for-Schleife

dates = [1982,1980,1973]   # Liste mit Jahreszahlen
anzahl = len(dates)        # die Funktion len() gibt die Anzahl der Elemente in der Liste an

for i in range(anzahl):    # für jede der 3 Jahreszahl zählt Python hier so ab: 0, 1, 2 und speichert den Wert jedesmal in i
    print(dates[i])        # wenn man ein Element aus der Liste wissen will verwenden man den Index in eckigen Klammern. dates[0] ist hier 1982
```

Der Code im Einzug wird `anzahl` mal ausgeführt, wobei jedes Mal der Wert von `i` um 1 erhöht wird. 

1.   zuerst ist i = 0
2.   dann ist i=1 und 
3.   zum Schluss ist i=2

Die ausgeführte Anweisung lautet dann: Schreibe (`print`) den Wert in der Liste beim Index `i`. Hier wird das zur Verdeutlichung dargestellt:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/LoopsForRange.gif" width="800" />

In diesem Beispiel können wir eine Zahlenreihe von 0 bis 7 ausdrucken:

```python
# Schleifen-Beispiel

for i in range(0, 8):
    print(i)
```

In Python können wir wie folgt direkt auf die Elemente in der Liste zugreifen: 

```python
# Schleifen-Beispiel, in einer Schleife eine Liste durchgehen

for year in dates:  
    print(year)   
```

Bei jeder Wiederholung verhält sich der Wert der Variablen `years` wie der Wert von `dates[i]` im ersten Beispiel:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/LoopsForList.gif" width="800">

Wir können die Elemente in einer Liste auch mit einer Schleife ändern:

```python
# Verwendung einer for-Schleife zum Ändern der Elemente in der Liste

quadrate = ['rot', 'gelb', 'grün', 'lila', 'blau']

for i in range(0, 5):
    print("Vor dem Ändern am Index ", i, 'steht dort',  quadrate[i])
    quadrate[i] = 'dunkelrot'
    print("Nach dem Ändern am Index ", i, 'steht dort',  quadrate[i])
```

##`while`-Schleifen##

Wie du sehen kannst, muss man bei der `for`-Schleife **vorher** wissen, wie oft man den Code durchlaufen möchte. Die Anzahl der Wiederholungen wird also im *Vor*hinein festlegt.

Was ist jedoch, wenn wir nicht wissen, nach wievielen Durchgängen wir die Schleife stoppen wollen? 

Wie können wir erreichen, dass ein Codeblock so lange ausgeführt wird, bis eine bestimmte Bedingung erfüllt ist? 

Diese Möglichkeit bietet uns die **`while`-Schleife**. Damit werden Anweisungen so lange ausgeführt **bis eine bestimmte Bedingung nicht mehr zutrifft**. Der Codeblock wird also so lange ausgeführt, bis die angegebene Bedingung einen **False**-Wert zurückgibt.

Nehmen wir an, wir möchten Schritt für Schritt die Liste `dates` auslesen und beim Jahr 1973 anhalten. Dann soll die Anzahl der Schritte ausgegeben werden, die wir gebraucht haben. 

Das kann man mit dem folgenden Codeblock erreichen:

```python
# Beispiel für eine While-Schleife

dates = [1982, 1980, 1973, 2000]   # eine Liste mit Jahreszahlen

i = 0                              # Die Integer-Variablen i und
year = 0                           # year werden erstellt und auf 0 gesetzt

while(year != 1973):               # Solange das in year gespeicherte Jahr nicht 1973 ist mache folgendes:
    year = dates[i]                # Speichere das Jahr am Index i in der Variablen year ab
    i = i + 1                      # i ist jetzt um 1 größer 
                                   # → beim nächsten mal wird in der Zeile darüber also das nächste Jahr in der Liste abgespeichert
    print(year)
                                   # keine Einrückung mehr, hier endet die Schleife
print("Es hat bis zum Verlassen der Schleife ", i ," Wiederholungen gebraucht.")
```

Eine while-Schleife wiederholt sich lediglich, bis die Bedingung im Argument nicht erfüllt ist. Was beim oberen Code passiert, wird in der folgenden Abbildung dargestellt:

##Beispiel1: Zahlenratespiel

Versuche zu verstehen, wie das folgende Programm funktioniert.

```python
import random                # Ein Modul mit dem man Zufallszahlen erzeugen kann

zufallszahl = random.randint(0,50)   # Zufallszahl von 0 bis 50

versuche = 0
weiter = True

while weiter:                # Unendliche Schleife, solange weiter=True
    
    versuche = versuche+1
    print("VERSUCH " + str(versuche))
    meinezahl = int(input("Wie lautet die gesuchte Zahl? "))
    
    if meinezahl == zufallszahl:
        print("Gratuliere! Gelöst mit " + str(versuche) + " Versuchen.")
        weiter = False
    elif meinezahl > zufallszahl:
        print("Deine Zahl ist zu groß!")
    else:
        print("Deine Zahl ist zu klein!")
```

##Beispiel2: Schere, Stein, Papier

Dieses Spiel kann man nicht nur einmal spielen. Wie wurde das gemacht?

Wenn du deine Auswahl triffst musst du entweder 1, 2 oder 3 eingeben. Was passiert, wenn man eine andere Zahl eingibt? Warum?

Warum kommt es zu einem Fehler, wenn man zB "Schere" eingibt?

```python
import random

#Einleitung
print("*****************************")
print("*  SCHERE | STEIN | PAPIER  *")
print("*****************************\n\n")

#Variablen
figuren = ("Schere", "Stein", "Papier")     #Tupel: eine Liste, die man nicht verändern kann; Werte durch Komma getrennt in runden ()
spielen = True

while spielen:

    #Spielerfigur auswählen
    spielerauswahl = 0
    while spielerauswahl not in (1,2,3):
        spielerauswahl = int(input("[1]Schere [2]Stein [3]Papier\n"))

    spielerfigur = figuren[spielerauswahl - 1]      #Listen beginnen bei 0!

    #Computerfigur auswählen
    computerfigur = figuren[random.randint(0,2)]

    #Sieger?
    print ("\n********************")
    print ("* Spieler:  " + spielerfigur + " *")
    print ("********************\n")


    print ("********************")
    print ("* Computer: " + computerfigur + " *")
    print ("********************\n")


    if spielerfigur == computerfigur:
        print("😐!!UNENTSCHIEDEN!!😐")
    elif ((spielerfigur == "Schere" and computerfigur == "Stein") or
          (spielerfigur == "Stein" and computerfigur == "Papier") or
          (spielerfigur == "Papier" and computerfigur == "Schere")):
        print("😒!!VERLOREN!!😒")
    else:
        print("😄!!GEWONNEN!!😄")

    #Nochmals spielen?
    entscheidung = ""
    while entscheidung not in ("j","n"):
        entscheidung = str.lower(input("\nNoch einmal spielen?  [j]Ja [n]Nein \n"))
    if entscheidung == "n":
        spielen = False

print("Bye bye!")
```
