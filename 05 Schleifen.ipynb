<h3 id="for">Was ist eine <code>for</code>-Schleife?</h3>

Eine <code>for</code>-Schleife ermöglicht es dir, einen Codeblock mehrmals auszuführen. Du könntest das z. B. verwenden, wenn du jedes Element in einer Liste ausgeben möchtest.    
Versuchen wir, eine <code>for</code>-Schleife zu verwenden, um alle Jahre auszugeben, die in der Liste <code>dates</code> enthalten sind:

Das kann folgendermaßen realisiert werden:

```python
# Beispiel einer for-Schleife

dates = [1982,1980,1973]
anzahl = len(dates)

for i in range(anzahl):
    print(dates[i])     
```

Der Code im Einzug wird <code>anzahl</code> mal ausgeführt, wobei jedes Mal der Wert von <code>i</code> um 1 erhöht wird. Die ausgeführte Anweisung lautet <code>print</code> den Wert in der Liste beim Index <code>i</code>, wie es hier zur Verdeutlichung dargestellt wird:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/LoopsForRange.gif" width="800" />

In diesem Beispiel können wir eine Zahlenreihe von 0 bis 7 ausdrucken:

```python
# Schleifen-Beispiel

for i in range(0, 8):
    print(i)
```

In Python können wir wie folgt direkt auf die Elemente in der Liste zugreifen: 

```python
# Schleifen-Beispiel, in einer Schleife eine Liste durchgehen

for year in dates:  
    print(year)   
```

Bei jeder Wiederholung verhält sich der Wert der Variablen <code>years</code> wie der Wert von <code>dates[i]</code> im ersten Beispiel:

<img src="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/PY0101EN/Chapter%203/Images/LoopsForList.gif" width="800">

Wir können die Elemente in einer Liste ändern:

```python
# Verwendung einer for-Schleife zum Ändern der Elemente in der Liste

quadrate = ['rot', 'gelb', 'grün', 'lila', 'blau']

for i in range(0, 5):
    print("Vor dem Quadrat ", i, 'ist',  quadrate[i])
    quadrate[i] = 'Gewicht'
    print("Nach dem Quadrat ", i, 'ist',  quadrate[i])
```

Auf den Index und die Elemente einer Liste können wir wie folgt zugreifen: 

```python
# Durchläuft die Liste und iteriert sowohl auf Index- als auch auf Elementwert.

quadrate=['rot', 'gelb', 'grün', 'lila', 'blau']

for i, quadrat in enumerate(quadrate):
    print(i, quadrat)
```

<h3 id="while">Was ist eine <code>while</code>-Schleife?</h3>

Wie du sehen kannst, wird die <code>for</code>-Schleife für einen *vor*bestimmten Verlauf von Wiederholungen benutzt, da man die Anzahl der Wiederholungen im *Vor*hinein festlegt. Was ist jedoch, wenn wir nicht wissen, wann wir die Schleife stoppen wollen? Was, wenn wir einen Codeblock so lange ausführen wollen, bis eine bestimmte Bedingung erfüllt ist? Die <code>while</code>-Schleife bietet die Möglichkeit Anweisungen so lange auszuführen bis eine bestimmte Bedingung nicht mehr zutrifft. Der Codeblock wird so lange ausgeführt, bis die angegebene logische Bedingung einen **False**-Wert zurückgibt.

Nehmen wir an, wir möchten durch die Liste <code>dates</code> iterieren ("*sie schrittweise abarbeiten*") und beim Jahr 1973 anhalten. Dann soll die Anzahl der Schritte ausgegeben werden. Dies kann mit dem folgenden Codeblock erfolgen:

```python
# Beispiel für eine While-Schleife

dates = [1982, 1980, 1973, 2000]

i = 0
year = 0

while(year != 1973):
    year = dates[i]
    i = i + 1
    print(year)

print("Es hat bis zum Verlassen der Schleife ", i ," Wiederholungen gebraucht.")
```

Eine while-Schleife wiederholt sich lediglich, bis die Bedingung im Argument nicht erfüllt ist, wie in der folgenden Abbildung dargestellt wird:
