<h2 id="creating">Erstellen einer Klasse</h2>

Jetzt werden wir eine Klasse Circle erstellen, aber zuerst importieren wir die Bibliothek <code>matplotlib.pyplot</code>, um Figuren und Formen zeichnen zu können: 

```python
# Import der Bibliothek

import matplotlib.pyplot as plt
%matplotlib inline  
```

Zur Erstellung einer eigenen Klasse musst du mit dem Schlüsselwort <code>class</code> beginnen und dann den Namen der Klasse, wie in Abbildung 4 gezeigt, festlegen. In dieser Einheit wird die übergeordnete Klasse (**parent class**) immer *object* sein, dh unsere Klassen beinhalten beim Erstellen nur das allernötigste und sind ganz allgemein gehalten: 

<img src="http://march.nmssemriach.at/pythonkurs/bilder/Klassen.png" width="300" />

<i>Abbildung 4: Definition der Klasse Circle.</i>

Der nächste Schritt ist eine spezielle Methode <code>&#95;&#95;init&#95;&#95;</code> namens Konstruktor, mit der das Objekt initialisiert wird. Hier werden die Daten der übergebenen Attribute abgespeichert. 

Der Begriff <code>self</code> enthält alle Attribute des Objekts. Zum Beispiel gibt <code>self.color</code> den Wert des Attributs color und <code>self.radius</code> den Radius des Objekts zurück. 

Wir haben auch die Methode <code>add_radius()</code> mit dem Parameter <code>r</code>, die den Wert von <code>r</code> zum Attribut Radius addiert. Um auf den Radius zuzugreifen, verwenden wir die Syntax <code>self.radius</code>. Die gekennzeichnete Syntax ist in Abbildung 5 zusammengefasst:

<img src="http://march.nmssemriach.at/pythonkurs/bilder/ClassesCircle.png" width="600" />

<i>Abbildung 5: Beschreibung der Syntax des Objekts Circle.</i>

Das eigentliche Objekt ist unten abgebildet. Wir fügen die Methode <code>drawCircle</code> ein, um das Bild eines Kreises anzuzeigen. Wir setzen den Standardradius auf 3 und die Standardfarbe auf blau:

```python
# Erstellen der KLasse Circle

class Circle(object):
    
    # Konstruktor
    def __init__(self, radius=3, color='blue'):
        self.radius = radius
        self.color = color 
    
    # Methode
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)
    
    # Methode
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()  
```

<h2 id="rect">Die Rectangle Klasse</h2>

Wir erstellen eine Klasse Rectangle mit den Attributen length, width und color. Wir werden nur die Methode zum Zeichnen des Rechteck-Objektes hinzufügen:

```python
# Erstellung einer neuen Rectangle-Klasse zur Erstellung eines Rechteck-Objektes

class Rectangle(object):
    
    # Konstruktor
    def __init__(self, width=2, length=3, color='r'):
        self.length = length 
        self.width = width
        self.color = color
    
    # Methode
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.length ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

Wir erstellen das Objekt <code>SchmalesBlauesRechteck</code> vom Typ Rectangle. Seine Breite wird 2 und die Länge 10 sein, und die Farbe setzen wir auf blau:

```python
# Ein neues Objekt vom Typ Rectangle erstellen

SchmalesBlauesRechteck = Rectangle(2, 10, 'blue')
```

Wie bisher können wir auf die Attribute der Instanz der Klasse über die Punktnotation zugreifen:

```python
# Ausgabe der Länge über das Objektattribut length

SchmalesBlauesRechteck.length 
```

```python
# Ausgabe der Farbe über das Objektattribut color

SchmalesBlauesRechteck.color
```

Wir können das Objekt zeichnen:
